```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# class names from CIFAR-10
classes = ["plane", "car", "cat", "bird", "deer", "dog", "frog", "horse", "ship", "truck"]

# download and load CIFAR-10 dataset 
print("Downloading dataset")

transform = transforms.Compose([transforms.ToTensor()])
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

print(f"Training samples: {len(trainset)}")
print(f"Test samples: {len(testset)}")
print(f"Image shape: {trainset[0][0].shape}")
print(f"Number of classes: {len(classes)}")
```

```python
# show images 
def show_images(dataset, num_images=8):
    # 2 rows of 4 pics per row
    fig, axes = plt.subplots(2, 4, figsize=(12, 6))
    # multi dimensional array into a one dimensional array
    axes = axes.ravel()

    for i in range(num_images):
        img, label = dataset[i]

        # reorder dimentions from [C, H, W] to [H, W, C] so matplotlib can display the images
        img = img.permute(1,2,0)

        axes[i].imshow(img)
        axes[i].set_title(f"{classes[label]}")
        axes[i].axis("off")

    plt.tight_layout() # adjust subplot spacing to prevent overlap
    plt.show()

print("first 8 training images:")
show_images(trainset)
```
